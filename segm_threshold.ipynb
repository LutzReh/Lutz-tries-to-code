```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img1 = plt.imread("/workspaces/Lutz-tries-to-code/data/NIH3T3/im/dna-0.png")
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.imshow(img1, cmap="grey")
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
threshold = 0.10 #mehr oder weniger random ausgesucht

binary_img1 = np.zeros_like(img1)

for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        if img1[i,j]>threshold:
            binary_img1[i,j]=255
        else:
            binary_img1[i,j]=0

plt.figure()
plt.title("OG Image")
plt.imshow(img1, cmap="grey")
plt.colorbar()
plt.show()

plt.figure()
plt.title("Binarized Image")
plt.imshow(binary_img1, cmap="grey")
plt.colorbar()
plt.show()
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt_img1 = plt.imread("/workspaces/Lutz-tries-to-code/data/NIH3T3/gt/0.png")
plt.imshow(gt_img1, cmap="gray")
plt.colorbar()
```

Compute and print the Dice coefficient:

```python
# GT ist von 0 bis 1 also muss das geladene Bild normalisiert werden

binary_img1 = binary_img1 // 255


DC = (2*(np.sum(binary_img1*gt_img1))/(np.sum(binary_img1)+np.sum(gt_img1)))

print(DC)

plt.imshow(binary_img1, cmap='gray')
plt.title("Binary Image 1")
plt.colorbar()
plt.show()
plt.imshow(gt_img1, cmap='gray')
plt.title("Ground Truth Image")
plt.colorbar()
plt.show()
```
