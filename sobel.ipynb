**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    kernel = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    return convolve(img, kernel)
    
def sobel_v(img):
    kernel = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    return convolve(img, kernel)

def convolve(img, kernel):
    img_h, img_w = img.shape
    kernel_h, kernel_w = kernel.shape

    padding_h, padding_w = kernel_h // 2, kernel_w // 2

    padded_img = np.pad(img,((padding_h, padding_h),(padding_w,padding_w)), mode="constant")

    output = np.zeros_like(img)

    for i in range(img_h):
        for j in range(img_w):
            region = padded_img[i:i+kernel_h, j:j+kernel_w]
            output[i,j] = np.sum(region * kernel)
    
    return output

```

**[Task 2.3.]()** Test your implementations from above:

```python
gx = sobel_h(img)
gy = sobel_v(img)

plt.imshow(gx, cmap="grey")
plt.title("Sobel X Achse")
plt.colorbar()
plt.show()

plt.imshow(gy, cmap="grey")
plt.title("Sobel Y Achse")
plt.colorbar()
plt.show()
```

**[Task 2.4.]()** Compute the image gradient magnitude:

```python
image_gradient_magnitude = np.sqrt(gx**2 + gy**2)

plt.imshow(image_gradient_magnitude, cmap="grey")
plt.title("Gradient Magnitude von Lena:))), sieht wirklich supu dupi aus:))")
plt.colorbar()
plt.show()
```
